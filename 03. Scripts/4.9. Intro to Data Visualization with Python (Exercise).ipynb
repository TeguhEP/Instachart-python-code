```python
#create bar charts from 'pricel_label' column
bar_price_label = ords_prods_merge['price_label'].value_counts().sort_index().plot.bar()
plt.title('The Bar Chart of The Price Label', fontsize = 18, pad=20)
plt.ylabel('Frequency (in millions)', fontsize = 12)
plt.xlabel('Price Label',  fontsize = 12)

#save the bar chart from the age group variable
bar_price_label.figure.savefig(os.path.join(path, '04. Analysis', 'Visualizations', 'bar_price_label.png'))
```

```python
#create bar charts from 'frequency_flag' column
bar_frequency_flag = ords_prods_merge['frequency_flag'].value_counts().sort_index().plot.bar()
plt.title('The Bar Chart of The Regularity of The Users Ordering Behavior', fontsize = 18, pad=20)
plt.ylabel('Frequency (in millions)', fontsize = 12)
plt.xlabel('Customer Group by Regularity of Ordering Behavior ',  fontsize = 12)

#save the bar chart from the age group variable
bar_frequency_flag.figure.savefig(os.path.join(path, '04. Analysis', 'Visualizations', 'bar_frequency_flag.png'))
```

```python
#create bar charts from 'loyalty_flag' column
bar_loyalty_flag = ords_prods_merge['loyalty_flag'].value_counts().sort_index().plot.bar()
plt.title('The Bar Chart of The Loyalty of the IC Customers', fontsize = 18, pad=20)
plt.ylabel('Frequency (in millions)', fontsize = 12)
plt.xlabel('Customer Loyalty',  fontsize = 12)

#save the bar chart from the age group variable
bar_loyalty_flag.figure.savefig(os.path.join(path, '04. Analysis', 'Visualizations', 'bar_loyalty_flag.png'))
```

## Create 'busiest_day' column

```python
#check the frequency of every value in 'orders_day_of_week' column
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#create the 'Busiest_day' columns with for-loops
result=[]

for value in ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('least busy')
    else:
        result.append('regularly busy')
```

# Direction 2: Create the "busiest_days" column

```python
#check the frequency of every value in 'orders_day_of_week' column
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#create the 'Busiest_days' columns with for-loops
result=[]

for value in ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest days')
    elif value == 1:
        result.append('Busiest days')
    elif value == 3:
        result.append('Slowest days')
    elif value == 4:
        result.append('Slowest days')
    else:
        result.append('regularly busy')
```

```python
ords_prods_merge['busiest_days'] = result
```

# Direction 3: Check the values and note any observations

```python
#check the frequency of values
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

Observations: The "regularly busy" category is still the highest, but the number decreased from 22416875 to 12916111. This happened due to the merging of saturday (0) and sunday (1) into 'busiest days' and the merging of wednesday (3) and thursday (4) into 'slowest days'

# Direction 4: Create the "busiest_period_of_day" column

```python
#check the value of "order_hour_of_day" column
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Classification:
Most orders = 10 - 16;
Average orders = 7 - 9 & 17 - 23;
Fewest orders = 0 - 6

```python
ords_prods_merge['order_hour_of_day'].shape
```

```python
#create 'busiest_period_of_day' column using loc()
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <= 16) & (ords_prods_merge['order_hour_of_day'] >= 10), 'busiest_period_of_day'] = 'Most Orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <= 9) & (ords_prods_merge['order_hour_of_day'] >= 7), 'busiest_period_of_day'] = 'Average Orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <= 23) & (ords_prods_merge['order_hour_of_day'] >= 17), 'busiest_period_of_day'] = 'Average Orders'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <= 6) & (ords_prods_merge['order_hour_of_day'] >= 0), 'busiest_period_of_day'] = 'Fewest Orders'
```
